```python
import pandas as pd
import numpy as np
import glob
import os
import re

train_dataset_path = '../data/csv/train_data.csv'
train_data = pd.read_csv(train_dataset_path)
train_data.head()
```

```python
files_list = glob.glob("../data/wavs/wt_data" + '/**/*.wav', recursive=True)
files_list.sort()

entries = []

for i, file in enumerate(files_list):
    base_path = os.path.basename(file)
    no_ext = base_path.split(".")[0]
    no_num = re.sub(r'[0-9]+', '', no_ext)
    no_underscore = no_num.replace("_", " ")
    final_string = no_underscore.strip()
    #print(final_string)
    match = train_data[train_data["transcription"].str.contains(final_string, regex=False)].iloc[0]
    action = match['action']
    object = match['object']
    location = match['location']
    entries.append([file, "NA", final_string, action, object, location])

wt_data = pd.DataFrame(entries, columns = train_data.columns[1:])

wt_data.head()
```

```python
testwt_dataset_path = 'data/wt_data.csv'
wt_data.to_csv(testwt_dataset_path)
```
